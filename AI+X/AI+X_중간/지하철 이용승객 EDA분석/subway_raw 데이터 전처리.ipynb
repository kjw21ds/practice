```python
import pandas as pd
raw_df = pd.read_csv('rawfiles/CARD_SUBWAY_MONTH_202301.csv')
raw_df
```

```python
raw_df.info()
```

```python
raw_df = pd.DataFrame()
for month in range(1, 10):
    temp_df = pd.read_csv(f'rawfiles/CARD_SUBWAY_MONTH_20230{month}.csv')
    raw_df = pd.concat([raw_df, temp_df], axis=0)
for month in range(10, 13):
    temp_df = pd.read_csv(f'rawfiles/CARD_SUBWAY_MONTH_2023{month}.csv')
    raw_df = pd.concat([raw_df, temp_df], axis=0)

raw_df = raw_df.reset_index(drop=True)
raw_df
```

```python
raw_df.info()
```

```python
raw_df['사용일자'] = pd.to_datetime(raw_df['사용일자'], format='%Y%m%d')
raw_df.info()
```

```python
raw_df['요일'] = raw_df['사용일자'].dt.dayofweek
raw_df
```

```python
day_of_week = {
    0 : '월',
    1 : '화',
    2 : '수',
    3 : '목',
    4 : '금',
    5 : '토',
    6 : '일'
}
raw_df['요일'] = raw_df['요일'].map(day_of_week)
```

```python
raw_df['연월'] = raw_df['사용일자'].dt.strftime('%Y-%m')
raw_df['월일'] = raw_df['사용일자'].dt.strftime('%m-%d')
raw_df.head()
```

```python
raw_df['승하차총승객수'] = raw_df['승차총승객수'] + raw_df['하차총승객수']
raw_df = raw_df[['사용일자', '요일', '노선명', '역명', '승차총승객수', '하차총승객수', '승하차총승객수', '연월', '월일', '등록일자' ]]
raw_df.head()
```
