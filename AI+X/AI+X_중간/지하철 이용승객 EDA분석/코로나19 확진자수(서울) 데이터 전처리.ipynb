```python
# 만약 인코딩이 utf-8이 아니라면 변환 후 덮어쓰는 코드
import pandas as pd
import chardet

# 파일 경로 설정
original_file = f'rawfiles/코로나19 확진자 발생 현황(2020.01~2023.08).csv'

# 파일의 현재 인코딩 확인
with open(original_file, 'rb') as f:
    result = chardet.detect(f.read())
encoding = result['encoding']

# 원본 파일을 읽고 UTF-8로 저장
temp_df = pd.read_csv(original_file, encoding=encoding)
temp_df.to_csv(original_file, index=False, encoding='utf-8')
```

```python
import pandas as pd
raw_df = pd.read_csv('rawfiles/코로나19 확진자 발생 현황(2020.01~2023.08).csv')
raw_df
```

```python
raw_df.info()
```

```python
raw_df.isna().sum()
```

```python
raw_df['서울시 일자별 확진자'] = raw_df['서울시 일자별 확진자'].str.replace(',', '').astype(int)
```

```python
raw_df['서울시 누적 확진자'] = raw_df['서울시 누적 확진자'].str.replace(',', '').astype(int)
```

```python
raw_df['서울시 누적 사망자'] = raw_df['서울시 누적 사망자'].str.replace(',', '').astype(int)
```

```python
raw_df['서울시기준일'] = pd.to_datetime(raw_df['서울시기준일'], format='%Y-%m-%d')
raw_df.info()
```

```python
raw_df['연월'] = raw_df['서울시기준일'].dt.strftime('%Y-%m')
raw_df['월일'] = raw_df['서울시기준일'].dt.strftime('%m-%d')
raw_df.head()
```

```python
fpath = 'rawfiles/코로나19 확진자 발생 현황(2020.01~2023.08).csv'
raw_df.to_csv(fpath, index=False, encoding='utf-8-sig')
```
